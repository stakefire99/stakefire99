```python
import folium

# Create base map centered around average coordinates
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]

restaurant_map = folium.Map(location=map_center, zoom_start=5)

restaurant_map
```

```python
for index, row in df.head(100).iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(restaurant_map)

restaurant_map
```

```python
from folium.plugins import MarkerCluster

# Create base map
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=5)

# Create marker cluster
marker_cluster = MarkerCluster().add_to(restaurant_map)

# Add restaurant markers
for index, row in df.head(500).iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(marker_cluster)

restaurant_map
```

```python
chain_counts = df['Restaurant Name'].value_counts()

chains = chain_counts[chain_counts > 1]

print("Top Restaurant Chains:")
print(chains.head(10))
```

## 🔹 Task 1: Restaurant Ratings Analysis

### Objective
To analyze the distribution of restaurant ratings and identify common rating patterns.

### Observations
After removing unrated restaurants (rating = 0), the distribution of ratings follows a near-normal pattern. Most restaurants fall within the 3–4 rating range.

### Key Findings
- Most common rating range: **3–4 (4388 restaurants)**
- Very few restaurants have ratings below 2
- Average number of votes: **202.19**

### Insight
The majority of restaurants maintain moderate ratings, indicating a competitive market where most establishments perform at an average to good level. Extremely low or exceptionally high ratings are relatively rare.

---

## 🔹 Task 2: Cuisine Combination Analysis

### Objective
To identify common cuisine combinations and analyze their impact on ratings.

### Observations
North Indian cuisine is the most common, followed by combinations like North Indian & Chinese, Fast Food, and Cafe.

However, the highest-rated cuisine combinations are niche and specialized.

### Key Findings
- Most common cuisine: **North Indian**
- Highest-rated cuisines include:
  - Burger, Bar Food, Steak
  - American, Burger, Grill
  - Italian, Bakery, Continental

### Insight
While common cuisines dominate in quantity, niche and specialized cuisine combinations tend to achieve higher customer ratings, suggesting better quality or unique experiences.

---

## 🔹 Task 3: Geographic Analysis

### Objective
To visualize restaurant locations and identify geographic patterns.

### Observations
Restaurant locations form dense clusters in certain regions when plotted on a map using latitude and longitude.

### Key Findings
- High concentration of restaurants in specific areas
- Clear clustering in urban regions

### Insight
Restaurants are heavily concentrated in cities and commercial hubs, indicating that business activity is driven by population density and customer demand. Rural areas show significantly fewer establishments.

---

## 🔹 Task 4: Restaurant Chains Analysis

### Objective
To identify restaurant chains and analyze their ratings and popularity.

### Observations
Several well-known chains appear multiple times, indicating strong brand presence.

### Key Findings
- Most common chains:
  - Cafe Coffee Day (83 outlets)
  - Domino’s Pizza (79 outlets)
  - Subway (63 outlets)

- Highest-rated chains:
  - Talaga Sampireun (4.9)
  - Silantro Fil-Mex (4.85)

- Most popular (by votes):
  - Truffles
  - Joey’s Pizza
  - AB’s Absolute Barbecues

### Insight
Large chains dominate in terms of presence, but smaller or premium chains tend to achieve higher ratings. Popularity (votes) reflects customer engagement rather than overall quality.

---

## 📌 Overall Conclusion (Level 2)

The analysis reveals that the restaurant market is highly competitive, with most establishments maintaining moderate ratings. While common cuisines and large chains dominate in presence, niche cuisines and specialized restaurants tend to achieve higher customer satisfaction. Geographic clustering further indicates that restaurant businesses thrive in densely populated urban areas.

## 🔹 Task 1: Customer Sentiment Analysis

### Objective
To analyze customer sentiment using available textual data.

### Methodology
Since the dataset does not include detailed customer reviews, the "Rating text" column was used as a proxy for sentiment analysis. This column categorizes restaurants into qualitative labels such as Excellent, Very Good, Good, Average, and Poor.

### Observations
The distribution shows that most restaurants fall under "Average", followed by "Good" and "Not rated". Only a small number of restaurants are categorized as "Excellent" or "Poor".

### Insight
The sentiment distribution indicates that most restaurants provide an average customer experience. The presence of a large number of "Not rated" entries suggests limited customer feedback for many restaurants, which may affect their credibility and visibility.

---

## 🔹 Task 2: Votes vs Ratings Analysis

### Objective
To analyze the relationship between restaurant ratings and customer engagement (votes).

### Observations
The scatter plot shows a general upward trend where restaurants with higher ratings tend to receive more votes. However, the distribution is scattered, indicating variability.

### Key Findings
- Positive correlation between ratings and votes
- High-rated restaurants often have higher customer engagement
- Presence of outliers with extremely high votes
- Some mid-rated restaurants have low visibility

### Insight
While higher ratings contribute to increased popularity, they are not the only factor influencing customer engagement. External factors such as brand reputation, location, and marketing also play a significant role.

---

## 🔹 Task 3: Price Range Analysis

### Objective
To analyze how price range influences restaurant ratings.

### Observations
The boxplot shows that median ratings increase gradually with price range. Higher-priced restaurants tend to have better ratings compared to lower-priced ones.

### Key Findings
- Higher price ranges (3 and 4) show higher median ratings
- Lower price ranges have more variability in ratings
- Outliers exist across all price categories

### Insight
There is a positive relationship between price and perceived quality, as higher-priced restaurants generally receive better ratings. However, price does not guarantee quality, as some expensive restaurants still receive lower ratings.

---

## 📌 Overall Conclusion (Level 3)

The advanced analysis highlights that customer perception, engagement, and pricing are interconnected but not solely dependent on one factor. While higher ratings and pricing generally correlate with better customer perception and engagement, other factors such as brand presence, location, and service quality significantly influence restaurant success.
